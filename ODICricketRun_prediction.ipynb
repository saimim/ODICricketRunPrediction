```python
count = 1
delivery_df = pd.DataFrame()
for index, row in matches.iterrows():
    if count in [75,108,150,180,268,360,443,458,584,748,982,1052,1111,1226,1345]:
        count+=1
        continue
    count+=1
    ball_of_match = []
    batsman = []
    bowler = []
    runs = []
    player_of_dismissed = []
    teams = []
    batting_team = []
    match_id = []
    city = []
    venue = []
    for ball in row['innings'][0]['1st innings']['deliveries']:
        for key in ball.keys():
            match_id.append(count)
            batting_team.append(row['innings'][0]['1st innings']['team'])
            teams.append(row['info.teams'])
            ball_of_match.append(key)
            batsman.append(ball[key]['batsman'])
            bowler.append(ball[key]['bowler'])
            runs.append(ball[key]['runs']['total'])
            city.append(row['info.city'])
            venue.append(row['info.venue'])
            try:
                player_of_dismissed.append(ball[key]['wicket']['player_out'])
            except:
                player_of_dismissed.append('0')
    loop_df = pd.DataFrame({
            'match_id':match_id,
            'teams':teams,
            'batting_team':batting_team,
            'ball':ball_of_match,
            'batsman':batsman,
            'bowler':bowler,
            'runs':runs,
            'player_dismissed':player_of_dismissed,
            'city':city,
            'venue':venue
        })
    delivery_df = delivery_df.append(loop_df)
```

```python
df1['current_score'] = df1.groupby('match_id').cumsum()['runs']
```

```python
df1
```

```python
df1['over'] = df1['ball'].apply(lambda x:str(x).split(".")[0])
df1['ball_no'] = df1['ball'].apply(lambda x:str(x).split(".")[1])
df1
```

```python
df1['balls_bowled'] = (df1['over'].astype('int')*6) + df1['ball_no'].astype('int')
```

```python
df1['balls_left'] = 300 - df1['balls_bowled']
df1['balls_left'] = df1['balls_left'].apply(lambda x:0 if x<0 else x)
```

```python
df1
```

```python
df1['player_dismissed'] = df1['player_dismissed'].apply(lambda x:0 if x == '0' else 1)
df1['player_dismissed'] = df1['player_dismissed'].astype('int')
df1['player_dismissed'] = df1.groupby('match_id').cumsum()['player_dismissed']
df1['wickets_left'] = 10-df1['player_dismissed']
```

```python
df1
```

```python
df1['crr'] = (df1['current_score']*6)/(300 - df1['balls_left'])
```

```python
df1
```

```python
groups = df1.groupby('match_id')

match_ids = df1['match_id'].unique()
last_five = []
for id in match_ids:
  last_five.extend(groups.get_group(id).rolling(window = 90).sum()['runs'].values.tolist())
```

```python
df1['last_five'] = last_five
```

```python
df1
```

```python
final_df1 = df1.groupby('match_id').sum()['runs'].reset_index().merge(df1,on = 'match_id')
```

```python
x = final_df1.drop(['runs_x'],1)
y = final_df1.runs_x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = .2,random_state = 42)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse = False,drop = 'first'),['batting_team','bowling_team','city'])
]
,remainder='passthrough')
```

```python
pipe = Pipeline(steps = [
    ('step1',trf),
    ('step2',StandardScaler()),
    ('step3',XGBRegressor(n_estimators = 1500,learning_rate = 0.2,max_depth = 12, random_state = 1))
])
```

```python
pipe.fit(x_train,y_train)
y_pred = pipe.predict(x_test)
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```

```python
pickle.dump(pipe,open('ODIpipe.pkl','wb'))
```
